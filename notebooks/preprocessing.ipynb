```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset = pandas.read_csv("../datasets/raw_data/dataset.csv")
```

```python
dataset.info()
```

```python
not_null_features = []
for feature in dataset.columns:
    null_prop = dataset[feature].isnull().sum() / dataset.shape[0]
    if null_prop < 0.9:
        not_null_features.append(feature)
    print('feature: %s; null_prop: %s' % (feature, str(null_prop)))
```

```python
def add_time_series_features(dataset: pandas.DataFrame):
    dataset['month'] = pandas.to_datetime(dataset['report_date']).dt.month
    dataset['weekday'] = pandas.to_datetime(dataset['report_date']).dt.weekday

add_time_series_features(dataset)
```
